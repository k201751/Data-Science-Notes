6. Highest and Lowest Purchase Prices.

```python
highest = ecompurchase.loc[(ecompurchase['Purchase Price'] == ecompurchase.loc[:]['Purchase Price'].max())]
lowest = ecompurchase.loc[(ecompurchase['Purchase Price'] == ecompurchase.loc[:]['Purchase Price'].min())]
print('\nMax purchase is: ----------------------\n',highest)
print('\nMin purchase is: ----------------------\n', lowest)
```

7. Average Purchase Price

```python
sum1 = ecompurchase.loc[:,'Purchase Price'].sum()
count1 = ecompurchase.loc[:,'Purchase Price'].count()
avg1 = sum1/count1
print(avg1)
```

13. How many people purchase during the AM and how many people purchase during PM?

```python
queryam = ecompurchase.loc[(ecompurchase.loc[:]['AM or PM'] == 'AM')].count()
querypm = ecompurchase.loc[(ecompurchase.loc[:]['AM or PM'] == 'PM')].count()
countam = queryam['AM or PM']
countpm = querypm['AM or PM']
print('People in AM: ', countam, '\nPeople in PM: ', countpm)
```
