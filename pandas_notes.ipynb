# **Test Code**

```python
import pandas as pd

# Create a DataFrame
data = {
    'Name': ['Ali', 'Sara', 'John', 'Emma'],
    'Age': [25, 22, 28, 24],
    'City': ['Lahore', 'Karachi', 'New York', 'London'],
    'Score': [90, 85, 88, 92]
}

df = pd.DataFrame(data)

# Display the DataFrame
print("Original DataFrame:")
print(df)

# Basic operations
print("\nSummary statistics:")
print(df.describe())  # Summary statistics for numeric columns

print("\nFilter rows where Score > 85:")
print(df[df['Score'] > 85])  # Filter rows with Score greater than 85

print("\nAdd a new column (Pass/Fail):")
df['Pass/Fail'] = df['Score'].apply(lambda x: 'Pass' if x >= 60 else 'Fail')
print(df)

print("\nSort the DataFrame by Age:")
print(df.sort_values(by='Age'))  # Sort by the 'Age' column
```
